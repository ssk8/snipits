```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data1 = pd.read_csv('ti1.txt', sep=" ", header=None)
data1.columns = ["ti_time", "ti_temp"]


data2 = pd.read_csv('ss1.txt', sep=" ", header=None)
data2.columns = ["ss_time", "ss_temp"]

data = pd.DataFrame()

data['ti_time_min'] = (data1["ti_time"]-data1["ti_time"][0])/60
data['ti_temp'] = data1["ti_temp"]
data['ss_time_min'] = (data2["ss_time"]-data2["ss_time"][0])/60
data['ss_temp'] = data2["ss_temp"]
```

```python
fig,ax = plt.subplots()

ax.plot(data["ti_time_min"], data['ti_temp'], label="ti")
ax.plot(data["ss_time_min"], data['ss_temp'], label="ss")

t = "240ml water in 300ml cup with lid, preheated for ~30 sec before fill"
plt.text(0, 0, t, ha='left', wrap=True)

ax.set_ylim([35,93])
ax.set_xlabel("time (min)")
plt.xticks(range(0, 185, 30))
plt.axhline(y=57)
ax.set_ylabel("temp (C)")
ax.legend(loc='best')
```
